# 4. load the image in grayscale

```python
image = cv2.imread('lina2.png', cv2.IMREAD_GRAYSCALE)
```

# 5. compute the histogram of the original image

```python
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
```

# 6. Apply the histogram equalization

```python
equalized_image = cv2.equalizeHist(image)
```

# 7. Compute the histogram of the equalized image

```python
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.plot(hist_original, color='black')
plt.title("Histogram of Original Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.plot(hist_equalized, color='black')
plt.title("Histogram of Equalized Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## Effect of Histogram Equalization
* Before equalization, the histogram may be concentrated in a narrow intensity range (low contrast).
* After equalization, the pixel values are spread across the full range (higher contrast).

```python
image = cv2.imread('barbara.jpg', cv2.IMREAD_GRAYSCALE)
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
equalized_image = cv2.equalizeHist(image)
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.plot(hist_original, color='black')
plt.title("Histogram of Original Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.plot(hist_equalized, color='black')
plt.title("Histogram of Equalized Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
